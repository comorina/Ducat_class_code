```python
from tkinter import *
from tkinter import messagebox
win=Tk()
win.state('zoomed')
win.resizable(width=False,height=False)
header_frm=Frame(win)
header_frm.configure(bg='pink')
header_frm.place(x=0,y=0,relwidth=1,relheight=.2)

title_lbl=Label(header_frm,text="Student Mgt System",bg='pink',font=('',50,'bold','underline'))
title_lbl.pack()

def main_body():
    def reset():
        user_entry.delete(0,"end")
        pass_entry.delete(0,"end")
        user_entry.focus()
    
    def login():
        u=user_entry.get()
        p=pass_entry.get()
        if(len(u)==0 or len(p)==0):
            messagebox.showwarning('validation',"Username/Password can't be empty")
        else:
            if(u=='admin' and p=='admin'):
                frm.destroy()
                login_body()
            else:
                messagebox.showerror('validation',"Invalid Username/Password")
    frm=Frame(win)
    frm.configure(bg='powder blue')
    frm.place(x=0,rely=.2,relwidth=1,relheight=.8)
    user_lbl=Label(frm,text="Username:",font=('',25,'bold'),bg='powder blue')
    user_lbl.place(relx=.3,rely=.2)

    user_entry=Entry(frm,font=('','20'),bd=7)
    user_entry.focus()
    user_entry.place(relx=.45,rely=.2)
    
    pass_lbl=Label(frm,text="Password:",font=('',25,'bold'),bg='powder blue')
    pass_lbl.place(relx=.3,rely=.35)

    pass_entry=Entry(frm,font=('','20'),bd=7,show='*')
    pass_entry.place(relx=.45,rely=.35)
    
    login_btn=Button(frm,text="login",font=('','20'),bd=7,command=login)
    login_btn.place(relx=.4,rely=.5)
    
    reset_btn=Button(frm,text="reset",font=('','20'),bd=7,command=reset)
    reset_btn.place(relx=.5,rely=.5)

    
def login_body():
    def logout():
        frm.destroy()
        main_body()
        
    frm=Frame(win)
    frm.configure(bg='powder blue')
    frm.place(x=0,rely=.2,relwidth=1,relheight=.8)
    wel_lbl=Label(frm,text="Welcome,admin",font=('',20),bg='powder blue')
    wel_lbl.place(relx=0,rely=0)
    
    logout_btn=Button(frm,text="logout",font=('',20),bd=7,command=logout)
    logout_btn.place(relx=.92,rely=0)
    
    reg_btn=Button(frm,text="Register Student",font=('',20),bd=7,width=15)
    reg_btn.place(relx=.4,rely=.1)
    
    find_btn=Button(frm,text="Find Student",font=('',20),bd=7,width=15)
    find_btn.place(relx=.4,rely=.3)
    
    dep_btn=Button(frm,text="Deposit Fee",font=('',20),bd=7,width=15)
    dep_btn.place(relx=.4,rely=.5)
    
    course_btn=Button(frm,text="Add Course",font=('',20),bd=7,width=15)
    course_btn.place(relx=.4,rely=.7)
main_body()
win.mainloop()
```
